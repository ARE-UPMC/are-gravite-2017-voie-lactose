## Position et calcul de forces des étoiles :
Positionne des points avec une position et une masse aléatoire (cohérente) dans un cercle représentant la Voie Lactée.

Par la suite ce programme détermine les forces extérieures qui lui sont appliquées^.

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar  7 19:12:13 2017

@author: benjamin

rajouter une variable faisant varier la masse
"""

from pylab import *
from math import*
import numpy as np
from random import randint


#Masse Soleil = 1,9891e30

Xmetre = []
Ymetre = []

Xkparsec = []
Ykparsec = []

Masse_Etoile=[]

def points_de_base(n):
    #Rayon de la Voie Lactée = 4.730365e20m
    #Rayon de la Voie Lactée = 15.33007kparsec
    Rayon_en_kparsec = 15
    
    
    
    Masse_soleil = 1.9891e30
    Masse_etoile_min = 3.51918e5
    Masse_etoile_max = 300*Masse_soleil
       
    
    circle=plt.Circle((0,0),Rayon_en_kparsec, color = 'r',fill = False)
    
    ax = plt.gca()
    ax.cla()     
    
    for i in range(n):
        
        xkparsec = randint(-Rayon_en_kparsec, Rayon_en_kparsec) 
        ykparsec = randint(-Rayon_en_kparsec, Rayon_en_kparsec)
        m = randint(Masse_etoile_min,Masse_etoile_max)
        
        if abs((sqrt((xkparsec**2)+(ykparsec**2)))) <= Rayon_en_kparsec :
            
            xmetre = (3.086e16)*xkparsec
            ymetre = (3.086e16)*ykparsec
            
            scatter([xkparsec],[ykparsec], 20, color='blue')
            Xkparsec.append(xkparsec)
            Ykparsec.append(ykparsec)
            Xmetre.append(xmetre)
            Ymetre.append(ymetre)
            Masse_Etoile.append(m)

    scatter([0],[0], 200, color='red', label='CDVL')
    scatter([xkparsec],[ykparsec], 20, color='blue', label='Etoiles') 
    legend(loc='upper left')
    
    xlim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    ylim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    
    ax.add_artist(circle)
    
    plt.show()
    
    print("Coordonnées des poins (en mètre puis en kparsec)")
    print(" ")
    print(Xmetre)
    print("------------------------------------------")
    print(Xkparsec)
    print(" ")
    print(" ")
    print(" ")
    print("Masses des étoiles (en kg)")
    print(" ")
    print(Masse_Etoile)
    
    
    return(Xmetre,Ymetre,Xkparsec,Ykparsec,Masse_Etoile)

def nombres_interactions():
    
    #Forces = dict() 
    #Contient la somme des forces appliquées à une étoile
    Forces = []
    
    x, y, xkparsec, ykparsec, masses_etoiles = points_de_base(5)
    #rlim=9.460731e+16
    #rlim=9.460731e+2
    
    #Constante de gravitaion
    G = 6.67e-11
    
    #Ajout des propriétés du centre de la galaxie
    Masse_centre_VoieLactee = 1.9891e46
    x_Centre_VoieLactee = 0    
    y_Centre_VoieLactee = 0 
    
    x.append(x_Centre_VoieLactee)
    y.append(y_Centre_VoieLactee)
    masses_etoiles.append(Masse_centre_VoieLactee)

    for i in range(len(x)):

        Somme_Forces = (0,0)        
        Fx = 0
        Fy = 0
        
        x1 = x[i]
        y1 = y[i]
        m1 = masses_etoiles[i]
        
        print(" ")
        print(" ")
        print('Angle du vecteur position du point' )
        print(i)
        
        Compteur_de_force = 0
        
        for j in range(len(x)): 
                        
            
            print('par rapport au point' )
            print(j)
            
            Force = 0
            
            x2 = x[j]
            y2 = y[j]
            m2 = masses_etoiles[j]
                            
            xA = x1 - x2
            yA = y1 - y2
                            
            distance_entre_etoile = sqrt((xA**2)+(yA**2))
            u = sqrt((xA**2)+(yA**2))
            #v-> est le vecteur unitaire de l'axe des abscisses
            #v->(1,0)
            v = 1
            u_v = xA*1
            
            teta = 0            
            
            if u*v == 0:
                teta = 0
            else :
                costeta = (u_v)/(u*v)
                teta = math.acos(costeta)
                
            if distance_entre_etoile != 0:
                Force = G*m1*m2/(distance_entre_etoile**2)
                
            print(teta)
            print('Compteur de force calculées pour ce point')
            print(Compteur_de_force)
            #Coordonées de la force
            Fx = Fx + Force*(math.cos(teta))
            Fy = Fy + Force*(math.sin(teta))
            
            Somme_Forces = (Fx, Fy)    
                
        Forces.append(Somme_Forces)
        print(" ")
        print(" ")
        print('Nombre final de forces appliquées sur le point')
        print(Compteur_de_force)
        print(" ")
        print(" ")
    print("")
    print("")
    print("Coordonnées (x,y) des forces appliquées à chaque point :")
    print (Forces)    
    return (Forces)
                
nombres_interactions()
```
