### L'objectif principal de la première partie de la programmation était de s'inspiré du site internet de l'Observatoire Européen Austral ESO pour créer des fonctions permettant de reprouidre les bras spiraux de la Voie Lactée et mettre en place la courbe de rotaion de notre galaxie. Sur le site internet de l'Observatoire Européen Austral ESO, on pouvait créer notre propre courbe de rotation de l'univers à partir des données qu'on pouvait observer via leur satelites, et on pouvait aussi reproduire certains bras spatiaux de la Voie Lactée via la positions de nuages de gaz.



```python
import math
def fonction_R(l):
    LR=[]
    r = (2*math.pi)/360
    for i in range(0,l):
        temp= 8.5*sin(i*r)
        if temp >=0:
            LR.append(8.5*sin(i*r))
    return LR
```

## Modélisation de la trajectoire des étoiles de la Voie Lactée

### En quoi consiste le programme ?
Le programme va générer aléatoirement des étoiles ayant des coordonnées aléatoires, sur un graphique. Ces étoiles seront présentes à l'intérieur de la Voie Lactée dont ses limites sont représentées par un cercle rouge sur le graphique.
En calculant la résultante des forces appliquées par tous les corps sur chaque étoile, le programme déterminera de nouvelles coordonnées pour chaque étoile modélisant ainsi leur déplacement. Ce processus ce répètera autant de fois que voulu.

### Analyse du programme

#### Différentes fonctions
Le programme est séparé en 4 fonctions.

La première défini la figure initiale à afficher. Elle génère les étoiles qu'elle représente par des points. Elle place ces points sur un graphique. Tous ces points sont compris dans les dimensions de la Voie Lactée dont ses limites sont représentées par un cercle rouge. Cette fonction place aussi au centre de la figure un point rouge représentant le centre de la Voie Lactée. Les distances sur le schéma sont mises à l'échelle et sont en kpc. Le programme va stocker dans une liste les coordonnées de chaque point. Elle va aussi convertir ces coordonnées en kpc, en m afin de faciliter les calculs plus tard. Toutes les coordonnées sont stockées dans 2 listes (une pour l'axe des x et une autre pour l'axe des y). De plus, la fonction va finir par définir une masse à chaque étoile. Si il y a beaucoup d'étoile, la masse de ces dernières sera la masse totale de la galaxie à laquelle on aura soustrait celle du centre de cette dernière, le tout divisé par le nombre d'étoiles présentes sur le schéma.

La deuxième fonction détermine la masse de la matière noire agissant sur chaque étoile en fonction de sa position sur le graphique.

La troisième fonction quand-à elle récupère les listes des coordonnées de chaque points, ainsi que leur masse. A partir de ces informations, cette fonction va calculer pour chaque point, la force d'attraction exercée par chaque autre point sur ce dernier. Ces forces seront additionnées et ceci donnera la résultante (selon l'axe des abscisses et des ordonnées) des forces appliquées à chaque point. Afin de calcul l'orientation de la résultante des forces appliquée sur chaque point, nous déterminons l'ange du vecteur de la force exercée par un corps sur un autre corps. Ceci nous permet de déterminer les composantes de chaque force selon l'axe des abscisses et celui des ordonnées. Pour y parvenir, nous utilisons la formule cos(teta)=(u->.v->)/uv, où u-> est le vecteur de la force et v-> est le vecteur unitaire de l'axe des abscisses, et uv est le produit des normes de chacun de ces vecteurs. Par la suite, il est facile de déterminer l'angle, sachant qu'il faut déterminer son signe en s'aidant du sinus (comme nous pouvons l'observer dans le programme).

La quatrième et dernière fonction est la fonction qui actualise les coordonnées de chaque point. Ceci est effectué en déterminant les équations de la trajectoire de chaque point, à l'aide de la résultante des forces appliquée à chacun de ces derniers.Cette fonction fait les calculs grâce aux coordonnées exprimées en m, afin de les placer à nouveau sur le graphique, elle les converti en kpc.
