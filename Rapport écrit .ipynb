## Modélisation de la trajectoire des étoiles de la Voie Lactée 

### En quoi consiste le programme ? 

Le programme va générer aléatoirement des étoiles ayant des coordonnées aléatoires, sur un graphique. Ces étoiles seront présentes à l'intérieur de la Voie Lactée dont ses limites sont représentées par un cercle rouge sur le graphique. 

En calculant la résultante des forces appliquées par tous les corps sur chaque étoile, le programme déterminera de nouvelles coordonnées pour chaque étoile modélisant ainsi leur déplacement. Ce processus se répètera autant de fois que voulu. 

### Analyse du programme 

#### Le commencement 

Le programme débute par l’initialisation de variables qui seront utilisées par plusieurs fonctions. Tout d’abord, la variable n correspond au nombre d’étoile que nous voulons intégrer au processus et tours est le nombre de tours que le programme va faire. Ensuite les listes qui contiendront les coordonnées des points (exprimées en mètre et en kpc) sont initialisées. Ensuite les valeurs clefs comme la masse solaire, la masse de la matière noire dans la Voie Lactée, celle du trou noir présent au centre de la galaxie et l’intervalle de temps utilisé dans le programme sont introduites. Ensuite s’en suivent 5 fonctions distinctes. 


##### I) Initialisation
La première fonction, la fonction initialisation a pour but de placer des points sur un graphique. Ces points représentent les étoiles de la Voie Lactée. La fonction va aussi mettre en place un cercle rouge sur le graphique, de façon à délimiter visuellement les bords de la Voie Lactée. Tous les points générés sont donc à l’intérieur du cercle rouge sur le graphique. A chaque point (et donc étoile) créé, il lui est associé des coordonnées (x, y) exprimées en mètres et en kpc, ainsi qu’une masse et une vitesse initiale. 

Tout d’abord, la fonction détermine la masse disponible dans la Voie Lactée que l’on peut attribuer aux étoiles. Pour cela, la fonction soustrait à la masse totale de la Voie Lactée, la masse du trou noir présent en son centre, ainsi que la masse totale de la matière noire présente dans la galaxie. La masse totale obtenue est divisée par le nombre approximatif d’étoiles qu’il y a dans la Voie Lactée (environ 300 milliards). Cette masse correspond à la masse associée à chaque point. Tous les points auront donc la même masse. 

Ensuite, un cercle rouge de rayon 15kpc est affiché sur le graphique, il affiche visuellement les limites de la Voie Lactée (qui a donc pour rayon 15kpc). 

Après avoir effectué cela, le programme ajoute aux listes indiquées précédemment, les coordonnées du centre de la Voie Lactée (x = 0, y = 0). Ensuite, la partie chargée de générer des points ayant des positions aléatoires entre en jeu. La fonction va créer des positions (x et y) aléatoires dans un carré ayant pour côté la longueur de la galaxie. Cependant avec ce processus, des points peuvent se trouver dans le carré mais en dehors du cercle rouge. Pour y remédier, la distance entre le point et le centre de la galaxie (centre du graphique) est calculée grâce à la formule u = sqrt((x^^2)+(y^^2)). Si la distance est supérieure au rayon de la galaxie, le point n’est pas pris en compte, sinon il est affiché sur le graphique et ses coordonnées (en mètre et en kpc) sont ajoutées aux listes prévues à cet effet. De plus une vitesse initiale lui est attribuée dans une liste. Nous avons orienté cette vitesse selon l’axe –Uteta (dans la base polaire) si l’angle entre le vecteur reliant le centre de la galaxie à l’étoile 

et le vecteur unitaire de coordonnées (1,0,0) est négatif, sinon la vitesse est orientée selon l’axe Uteta. L’angle est calculé par le 2ème fonction. 

Ensuite, nous mettons en place la légende et les limites du graphiques. La fonction retourne les 4 listes contenant les coordonnées des points, les listes comportant les composantes (x, y) des vitesses initiales, ainsi que la valeur de la masse associée aux étoiles. 


##### II) Calcul angle 

La deuxième fonction, la fonction calcul_angle a pour objectif de calculer l’angle entre 2 vecteurs. Cette fonction est utile pour calculer l’orientation des forces appliquées entre les étoiles par exemple. Elle prend en compte 8 paramètres correspondant aux coordonnées de 4 points, permettant ainsi de créer les 2 vecteurs dont nous avons besoin. 

Tout d’abord, la fonction va calculer la longueur des 2 vecteurs grâce à la formule citée dans la partie précédente (avec x = x2-x1 et y = y2-y1 par exemple). Ensuite elle va calculer le produit scalaire entre ces 2 vecteurs. : u.v = x*x’+y*y’. 

Les cas créant des erreurs sont gérés. Par exemple, lorsque la longueur d’un vecteur est nulle. Par la suite, le calcul de l’angle est effectué grâce à la formule cos (teta) = (u.v)/u*v où u.v est le produit scalaire des 2 vecteurs et u et v sont les longueurs des 2 vecteurs. Ensuite il est facile de déterminer la valeur de l’angle lorsqu’on possède celle de son cosinus. La valeur du sinus de l’angle est utilisée pour déterminer le signe de l’angle. En effet si le sinus est négatif alors l’angle est négatif, sinon il est positif. 

La fonction retourne la valeur de l’angle entre les 2 vecteurs, ainsi que la longueur de chacun des vecteurs. 


##### III) Calcul masse centre VL avec matière noire 

La troisième fonction a pour but, comme son intitulé l’indique, de calculer la masse de la matière noire agissant sur une étoile située à une distance donnée du centre de la galaxie. 

Pour cela, nous avons préalablement calculé la masse surfacique de la matière noire aux bords de la Voie Lactée en calculant l’aire approximative de la Voie Lactée qui divise la masse totale de la matière noire dans notre galaxie. Ensuite, la fonction en déduit une deuxième masse surfacique en fonction de la distance à laquelle un point est situé du centre la Voie Lactée avec une décroissance en 1/r de la première masse surfacique. r est la distance mentionnée précédemment. Grâce à cette nouvelle masse surfacique, nous en déduisons la masse de matière noire agissant sur un point présent à un endroit donné. La fonction retourne la somme de cette masse avec la masse du trou noir présent au centre de la Voie Lactée. Nous avons choisi d’additionner ces masses car nous modélisant l’action de la matière noire sur un corps comme si elle était totalement concentrée au centre de la galaxie. Ainsi, c’est comme si le programme retournait la masse totale concentrée au centre la Voie Lactée agissant sur une étoile. 


##### IV) Forces 

La quatrième fonction, la fonction force détermine la résultante des forces appliquées sur chaque étoile par les autres astres. Elle prend en compte 5 paramètres, les listes contenant les coordonnées métriques et en kpc des points, ainsi que la masse des étoiles. 

Tout d’abord la fonction crée un epsilon qui correspond à la longueur d’un point sur le graphique. Ensuite la fonction initialise la liste qui contiendra la valeur des forces. Celle des forces appliquées au centre de la Voie Lactée est ajoutée en premières valeurs de la liste. En effet, toutes les listes du programme ont leur donnée ordonnée par étoile. C’est-à-dire que toutes les données concernant une étoile occuperont le même emplacement dans toutes les listes. 

Ensuite 2 itérations sont effectuées afin de parcourir les listes des points afin de calculer les forces appliquées par toutes les étoiles sur chacune d’entre elle. Les composantes d’une force appliquée par un astre sur un autre est calculée grâce à la formule F = -(G*(m1m2)/(r^2))*Ur où G est la constante universelle de gravitation, m1 et m2 sont les masses des 2 astres, r est la distance entre ces 2 astres et Ur est le vecteur de coordonnées x = cos (teta) et y = sin (teta). La valeur de l’angle teta et la distance entre les 2 astres sont récupérées grâce à la fonction calcul_angle présentée précédemment. Grâce à tout ceci, nous créons les composantes (x, y) d’une force appliquée à une étoile que l’on additionne aux autres forces appliquées par les autres étoiles sur cette dernière qui ont été calculées précédemment. Une fois que nous avons calculé toutes les forces appliquées par toutes les étoiles sur une étoile, nous ajoutons les valeurs des composantes des forces sous forme de tuple à la liste des forces initialisée au début de la fonction. Ce processus est répété pour toutes les étoiles du graphique. 

La fonction retourne la liste des forces appliquées sur chacune des étoiles, ainsi que les 4 listes de coordonnées et la masse des étoiles. 


##### V) Update 

La dernière fonction du programme, la fonction update, a pour but de calculer les nouvelles coordonnées de chaque point (et donc de chaque étoile) en fonction des forces extérieures appliquées sur celles-ci.  

La fonction ne prend pas de paramètre en compte et commence en initialisant les dont nous aurons besoins pour la suite du processus. Elle est composée d’une itération afin de parcourir tous les points. S’il s’agit du premier tour, la fonction initialise les données (positions, vitesses initiales et masse) en récupérant les données retournées par la fonction initialisation. Ensuite ces données sont réinjectées dans la fonction forces qui retourne les composantes des forces extérieures appliquées sur chaque étoile. Grâce à ces composantes, il nous ai possible de déterminer l’équation de la trajectoire de chaque étoile et donc de déterminer sa position à l’instant t+delta t (la valeur de delta t a été initialisée au début du programme). Pour y parvenir, la fonction effectue une nouvelle itération sur la liste des forces et calcul les composantes (x, y) de l’accélération pour chaque point grâce à la formule ax = Fx/masse de l’étoile et ay = Fy/masse de l’étoile. Ensuite, nous calculons les coordonnées de l’étoile à l’instant t+delta t grâce à la formule x = (1/2)*(ax^2)+v_précédente*(delta t)+x_précédente. Ensuite nous calculons la nouvelle vitesse de l’étoile, nous mettons à jour toutes les listes, 

nous affichons les points à leur nouvelles coordonnées. Ce processus est répété autant de fois que désirez.

```python
from pylab import *
from math import*
import numpy as np
from random import randint

"""CONDITIONS INITIALES"""
n = 10
Tours = 5
"""--------------------"""

Xmetre = []
Ymetre = []

Xkparsec = []
Ykparsec = []

Masse_soleil = 1.9891e30
Masse_centre_VoieLactee = (3.5e6)*Masse_soleil
Rayon_en_kparsec = 15
Nbre_d_etoile_VoieLactee = 300e9

#Conversion kparsec en m :
#1kpc = 3.086e19m
kpc = 3.086e19

#Intervalle de temps en secondes (1 millioin d'années)
Dt = 0.25*3.1536e13

def Initialisation(n):
    '''int -> list[int]**2 * int
    Initiatilise les positions, la masse et la vitesse initiale de n étoiles dans
    la Voie Lactée de façon aléatoire'''
    
    #Rayon de la Voie Lactée = 4.730365e20m
    #Rayon de la Voie Lactée = 15.33007kparsec
    Rayon_en_kparsec = 15
    Masse_Voie_Lactee = (5.6e11) * Masse_soleil
    Masse_matiere_noire_totale = 0.83*Masse_Voie_Lactee
    Masse_Disponible = (Masse_Voie_Lactee - (Masse_centre_VoieLactee + Masse_matiere_noire_totale))/Nbre_d_etoile_VoieLactee 
    
    vx = []
    vy = []

    if n == 1:
        Masse_Etoile = randint(-Masse_soleil, 300*Masse_soleil)
    else:
        Masse_Etoile = Masse_Disponible / n       
    
    circle=plt.Circle((0,0),Rayon_en_kparsec, color = 'r',fill = False)
    
    ax = plt.gca()
    ax.cla()     
    
    #Ajout des coordonnées du centre de la Voie Lactée :
    x_Centre_VoieLactee = 0    
    y_Centre_VoieLactee = 0 
    v_Centre_VoieLactee = 0
    
    Xmetre.append(x_Centre_VoieLactee)
    Ymetre.append(y_Centre_VoieLactee)
    Xkparsec.append(x_Centre_VoieLactee)
    Ykparsec.append(y_Centre_VoieLactee)
    vx.append(v_Centre_VoieLactee)
    vy.append(v_Centre_VoieLactee)
    
    scatter([Xkparsec[0]],[Ykparsec[0]], 200, color='red',label='Centre VL')
    
    for i in range(n):
        
        xkparsec = randint(-Rayon_en_kparsec, Rayon_en_kparsec) 
        ykparsec = randint(-Rayon_en_kparsec, Rayon_en_kparsec)
        
        if abs((sqrt((xkparsec**2)+(ykparsec**2)))) <= Rayon_en_kparsec :
            
            xmetre = (kpc)*xkparsec
            ymetre = (kpc)*ykparsec
            
            if xkparsec != 0:
                
                scatter([xkparsec],[ykparsec], 20, color='blue')
            teta, u, v = calcul_angle(0,0,xmetre,ymetre,0,0,1,0)
            if ymetre < 0 :
                vx.append((6.2e5)*sin(teta))
                vy.append((6.2e5)*cos(teta))
            else :
                vx.append(-(6.2e5)*sin(teta))
                vy.append((6.2e5)*cos(teta))
                
            Xkparsec.append(xkparsec)
            Ykparsec.append(ykparsec)
            Xmetre.append(xmetre)
            Ymetre.append(ymetre)
    
    scatter([Xkparsec[1]],[Ykparsec[1]], 20, color='blue', label='Etoiles') 
    legend(loc='upper left')
    
    xlim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    ylim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    
    ax.add_artist(circle)
         
    return(Xmetre,Ymetre,Xkparsec,Ykparsec,Masse_Etoile, vx, vy)

def calcul_angle(x1, y1, x2, y2, x3, y3, x4, y4):
    """calcule la longueur de 2 vecteurs, leur produit scalaire
    et détermine l'angle entre ces 2 vecteurs"""
                
    u = sqrt(( (x2-x1)**2)+( (y2-y1)**2))
    v = sqrt(((x4-x3)**2)+((y4-y3)**2))
    uscalv = (x2-x1)*(x4-x3)+(y2-y1)*(y4-y3)
                
    teta = 0
    if u*v == 0:
        teta = 0
    else :
        costeta = (uscalv)/(u*v)
        teta = math.acos(costeta)
    if math.sin(teta)<0 :
        teta = -teta
    return (teta, u, v)
    
def calcul_masse_centre_VL_avec_matiere_noire(masse, distance):
    """int**2 -> int
    Retourne la masse de la matière noire agissan sur une particule"""
    Masse = masse
    Distance = distance
    
    #masse surfaciue de la matière noire aux bords de la Voie Lactée:
    romax = 0.02 #kg/m²
    #masse surfacique de la matière noire appliquée au point :
    ro = romax*(1/Distance)
    Masse_matiere_noire = ro*(math.pi)*(Distance**2)
    return (Masse_centre_VoieLactee + Masse_matiere_noire)

def Forces(Xm,Ym,Xpc,Ypc,masse):
    '''list[int]**4 * int -> list[int]**5 * int
    A partir des coordonnées de tous les points, calcule la résultante des forces appliquées
    à chaque point'''
    #précision (2*taille d'un point en metre):
    #taille d'un poin sur figure : 2mm
    #sur figure : 10 kpc en 2.3cm
    #Donc un point représente 869.6pc soi 2.6835856e22m
    epsilon = 2.68347826e19
    
    #Contient la somme des forces appliquées à une étoile
    Forces = []
    
    x = Xm
    y = Ym
    xkpc = Xpc
    ykpc = Ypc
    masse_etoile = masse
    
    #Constante de gravitaion
    G = 6.67e-11
    Forces.append((0,0))
    for i in range(1,len(x)):

        Somme_Forces = (0,0)        
        Fx = 0
        Fy = 0
        
        x1 = x[i]
        y1 = y[i]
        
        Compteur_de_force = 0
        
        for j in range(0,len(x)): 
            
            #list[int] liste les distances entre chaque étoile et le centre de la VL
            Distances = []
            #list[int] liste la masse de la matière noire agissant sur cette éoile
            Masse_MT=[]
            
            Force = 0
            
            x2 = x[j]
            y2 = y[j]
            teta, distance_entre_etoile, v = calcul_angle(x1,y1,x2,y2, 0,0,1,0)
                            
            if distance_entre_etoile != 0 and x2 != x[0]:
                Force = -G*(masse_etoile**2)/((distance_entre_etoile+epsilon)**2)
            
            elif distance_entre_etoile != 0 and x2 == x[0]:
                Masse = calcul_masse_centre_VL_avec_matiere_noire(masse, distance_entre_etoile)
                Force = -G*(masse_etoile*Masse)/((distance_entre_etoile+epsilon)**2)
                Distances.append(distance_entre_etoile)
                Masse_MT.append(Masse-Masse_centre_VoieLactee)
                
            Fx = Fx + Force*(math.cos(teta))
            Fy = Fy + Force*(math.sin(teta))
            
            Somme_Forces = (Fx, Fy)    
                
        Forces.append(Somme_Forces)
        
    Forces[0] = (0,0)
    return (Forces,Xmetre,Ymetre,Xkparsec,Ykparsec,masse_etoile,teta)
                
def update():
    
    PX = []
    PY = []

    v_actuelle = []
    vx_actuelle = []
    vy_actuelle = []

    x_actuelle = []
    y_actuelle = []
    
    p1x = []
    p1y = []

    for t in range(Tours):
        print("")
        print("nombre de tours")
        print(t)
        if t == 0 :
            #Initialisation des valeurs :
            X, Y, Xkpc, Ykpc, Masse, vx_actuelle, vy_actuelle = Initialisation(n)
            forces, x_actuelle, y_actuelle, Xkpc, Ykpc, masse, teta = Forces(X, Y, Xkpc, Ykpc, Masse)
                        
            colour = 'green'
                
        forces, x_actuelle, y_actuelle, Xkpc, Ykpc, masse, teta = Forces(x_actuelle, y_actuelle, Xkpc, Ykpc, Masse)
    
        if t == 1 :
            colour = 'brown'
        if t == 2 :
            colour = 'red'
        if t == 3 :
            colour = 'black'
        if t == 4 :
            colour = 'purple'
        if t == 5 :
            colour = 'pink'
        if t == 6 :
            colour = 'gray'
          
        for k in range(len(forces)) :
            
            Fx, Fy = forces[k]

            #calcul de l'accélération :
            ax = Fx/masse
            ay = Fy/masse
                      
            #Calcul de la position :
            Vx = vx_actuelle[k]
            Vy = vy_actuelle[k]
            if k == 0:
                px = 0
                py = 0
            else :
                px = (1/2)*ax*(Dt**2) + Vx*Dt + x_actuelle[k]
                py = (1/2)*ay*(Dt**2) + Vy*Dt + y_actuelle[k]
                        
            PX = (px / kpc)
            PY = (py / kpc)
            Vx = ax*Dt+vx_actuelle[k]
            Vy = ay*Dt+vy_actuelle[k]
            
            
            scatter([Xkpc[k]],[Ykpc[k]], 20, color=colour)
            plot(Fx, Fy, color="blue", linewidth=20, linestyle="-")
            
            #mise à jour des valuers
            x_actuelle[k] = px
            y_actuelle[k] = py
            Masse = masse            
            vx_actuelle[k] = Vx
            vy_actuelle[k] = Vy
            Xkpc[k] = PX
            Ykpc[k] = PY
                    
    xlim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    ylim(-Rayon_en_kparsec*1.4, Rayon_en_kparsec*1.4)
    
    plt.show()
update()
```
